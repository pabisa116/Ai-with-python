```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(-5, 5, 100)

# Define y values for each line
y1 = 2*x + 1
y2 = 2*x + 2
y3 = 2*x + 3

# Create the figure
plt.figure(figsize=(8, 6))

# Plot the lines with different line styles
plt.plot(x, y1, 'k-', label='y = 2x + 1')   # solid black line
plt.plot(x, y2, 'k--', label='y = 2x + 2')  # dashed black line
plt.plot(x, y3, 'k:', label='y = 2x + 3')   # dotted black line

# Add labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Lines y = 2x + 1, y = 2x + 2, y = 2x + 3')

# Add grid and legend
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Step 1: Read CSV file ===
# Assuming the file "weight-height.csv" has a header like:
# Gender,Height,Weight
data = np.genfromtxt('weight-height.csv', delimiter=',', skip_header=1)

# === Step 2: Extract length (height) and weight columns ===
# Height is in column 1, weight is in column 2
length_in = data[:, 1]
weight_lb = data[:, 2]

# === Step 3: Convert units ===
# 1 inch = 2.54 cm
# 1 pound = 0.45359237 kg
length_cm = length_in * 2.54
weight_kg = weight_lb * 0.45359237

# === Step 4: Calculate means ===
mean_length = np.mean(length_cm)
mean_weight = np.mean(weight_kg)

print(f"Mean length: {mean_length:.2f} cm")
print(f"Mean weight: {mean_weight:.2f} kg")

# === Step 5: Draw histogram of lengths ===
plt.figure(figsize=(7, 5))
plt.hist(length_cm, bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Student Lengths (cm)')
plt.xlabel('Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Import NumPy
import numpy as np

# Step 1: Define the matrix A
A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

# Step 2: Compute the inverse of A
A_inv = np.linalg.inv(A)
print("A^-1 =\n", A_inv)

# Step 3: Check A * A^-1
product1 = np.dot(A, A_inv)
print("\nA * A^-1 =\n", product1)

# Step 4: Check A^-1 * A
product2 = np.dot(A_inv, A)
print("\nA^-1 * A =\n", product2)

# Optional: Round the results to make it easier to read
print("\nRounded products (to 5 decimals):")
print("A * A^-1 =\n", np.round(product1, 5))
print("A^-1 * A =\n", np.round(product2, 5))
```
