```python
import numpy as np
import matplotlib.pyplot as plt

# List of sample sizes
sample_sizes = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

# Loop through each sample size
for n in sample_sizes:
    # Step 1: Simulate throwing two dice n times
    dice1 = np.random.randint(1, 7, size=n)
    dice2 = np.random.randint(1, 7, size=n)
    s = dice1 + dice2  # Sum of the two dice

    # Step 2: Compute histogram of sums from 2 to 12
    h, h2 = np.histogram(s, bins=range(2, 14))  # range(2,14) gives bins for 2 to 12

    # Step 3: Plot histogram using normalized frequencies
    plt.figure(figsize=(8, 4))
    plt.bar(h2[:-1], h / n, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of Dice Sums (n = {n})")
    plt.xlabel("Sum of Two Dice")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set number of dice throws
n = 10000

# Simulate dice throws
dice1 = np.random.randint(1, 7, size=n)
dice2 = np.random.randint(1, 7, size=n)
s = dice1 + dice2

# Compute histogram
h, h2 = np.histogram(s, bins=range(2, 14))

# Plot histogram
plt.figure(figsize=(8, 4))
plt.bar(h2[:-1], h / n, color='skyblue', edgecolor='black')
plt.title(f"Histogram of Dice Sums (n = {n})")
plt.xlabel("Sum of Two Dice")
plt.ylabel("Normalized Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
When the number of dice throws is small, the histogram looks uneven and random. As the number of throws increases, the distribution smooths out and converges toward the expected probabilities, forming a clear symmetric shape centered at 7.
```
